```python
# Group by 'iter' and extract the relevant cost columns
average_costs = dataset.groupby('iter').agg({
    'Heur Cost': 'first',
    'Local Cost': 'first',
    'Delay Cost Local': 'first',
    'Penalty Local': 'first',
    'Fuel Cost Local': 'first',
    'Handling Cost Local': 'first',
    'Waiting Cost Local': 'first',
    'Delay Cost Heur': 'first',
    'Penalty Heur': 'first',
    'Fuel Cost Heur': 'first',
    'Handling Cost Heur': 'first',
    'Waiting Cost Heur': 'first',
    'better':'first',
    'usedLocalSearch':'first'
}).reset_index()

# Costs for plotting
costs_for_plotting = [
    ('Local Cost','Heur Cost', 'Total Cost'),
    ('Delay Cost Local','Delay Cost Heur', 'Delay Cost'),
    ('Waiting Cost Local', 'Waiting Cost Heur', 'Waiting Cost'),
    ('Penalty Local', 'Penalty Heur', 'Penalty Cost'),
    ('Handling Cost Local', 'Handling Cost Heur', 'Handling Cost'),
    ('Fuel Cost Local', 'Fuel Cost Heur', 'Fuel Cost')
]

iterations = average_costs['iter']



# Plotting
for cost_local, cost_heur, cost_name in costs_for_plotting:
    plt.figure(figsize=(10, 5))

    # Calculate the relative difference as a percentage
    relative_difference = 100 * (average_costs[cost_local] - dataset[cost_heur]) / average_costs[cost_heur]

    # Plot the line
    plt.plot(iterations, relative_difference, label='Relative Difference (%)', marker='o')

    # Adding titles and labels
    plt.title(f'Relative Difference in {cost_name} Over Iterations')
    plt.xlabel('Iteration')
    plt.ylabel('Relative Difference (%)')
    plt.legend()

    # Show plot
    plt.show()
```

```python
import pandas as pd

# Dictionary to store the mean relative differences for each cost type
mean_relative_differences = {}

# Calculate mean relative differences
for cost_local, cost_heur, cost_name in costs_for_plotting:
    # Calculate the relative difference as a percentage
    relative_difference = (average_costs[average_costs['usedLocalSearch']==1][cost_local] - average_costs[average_costs['usedLocalSearch']==1][cost_heur])
    # Calculate the mean relative difference and store it in the dictionary
    mean_relative_differences[cost_name] = relative_difference.mean()

# Convert the dictionary into a pandas DataFrame
mean_relative_differences_df = pd.DataFrame(list(mean_relative_differences.items()), columns=['Cost Type', 'Mean Relative Difference'])

# Plotting
plt.figure(figsize=(10, 5))
mean_relative_differences_df.plot(x='Cost Type', y='Mean Relative Difference', kind='bar', legend=False)
plt.title('Mean cost Difference in Dollars')
plt.ylabel('Mean Relative Difference (%)')
plt.xlabel('Cost Type')
plt.xticks(rotation=45)
plt.show()
```

```python
# List of column pairs (local and heuristic costs) and their labels
cost_pairs = [
    ('Local Cost', 'Heur Cost', 'Total Cost'),
    ('Delay Cost Local', 'Delay Cost Heur', 'Delay Cost'),
    ('Waiting Cost Local', 'Waiting Cost Heur', 'Waiting Cost'),
    ('Penalty Local', 'Penalty Heur', 'Penalty'),
    ('Handling Cost Local', 'Handling Cost Heur', 'Handling Cost'),
    ('Fuel Cost Local', 'Fuel Cost Heur', 'Fuel Cost')
]

# Calculate the differences between local and heuristic costs for each type
cost_differences = pd.DataFrame()

for local, heur, label in cost_pairs:
    cost_differences[label] = all_datasets[all_datasets['usedLocalSearch']==1].groupby('inst_type')[local].mean() - all_datasets[all_datasets['usedLocalSearch']==1].groupby('inst_type')[heur].mean()
# Transpose the DataFrame to have inst_type as columns
cost_differences = cost_differences.transpose()
```
